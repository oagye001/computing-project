```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_excel('EDGARv5.0_FT2019_fossil_CO2_booklet2020.xls',sheet_name='fossil_CO2_totals_by_country')
dataset.head()
```

```python
dataset[1970].values
```

```python
uk = dataset.loc[dataset.country_name == 'United Kingdom'].values.flatten()
uk.tolist()
print(uk)
```

```python
europedf = pd.DataFrame.melt(dataset, var_name='Year', value_name = 'Emission total', id_vars = ['country_name'])
europedf.head()
```

```python
europe = ['Albania','Armenia','Austria','Azerbaijan','Belarus','Belgium','Bosnia and Herzegovina','Bulgaria','Croatia','Cyprus','Czechia','Denmark','Estonia','Finland','France','Georgia','Germany','Greece','HUngary','Iceland','Ireland','Italy','Kazakhstan','Latvia','Lithuania','Luxembourg','Malta','Moldova', 'Netherlands', 'North Macedonia','Norway','Poland','Portugal','Romania','Russia','Serbia','Slovakia','Slovenia','Spain','Sweden','Switzerland','Turkey','Ukraine','United Kingdom']
europedf['country_name'] = europedf['country_name'].astype(pd.CategoricalDtype(categories = europe,ordered = True))
europedf['country_name'].dtype
```

```python
europedf.dropna(inplace = True)
europedf.head(20)
```
